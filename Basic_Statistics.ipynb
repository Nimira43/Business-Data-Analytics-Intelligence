```python
#libraries
import pandas as pd
import seaborn as sns
```

```python
#dataset
df = pd.read_csv('Baseball.csv')
df.head()
```

##Mean

```python
#Mean of runs scored - rs
df.RS.mean()
```

```python
#Arizona data
df.loc[df.Team == 'ARI']
```

```python
#Mean of runs scored - rs - Arizona
df.loc[df.Team == 'ARI'].RS.mean()
```

```python
#Arizona - since 2005
df.loc[(df.Team == 'ARI') & (df.Year > 2005)]
```

```python
#Mean of runs scored - rs - Arizona - since 2005
df.loc[(df.Team == 'ARI') & (df.Year > 2005)].RS.mean()
```

```python
#Chicago data
df.loc[df.Team == 'CHC']
```

```python
#Mean of runs scored - rs - Chicago
df.loc[df.Team == 'CHC'].RS.mean()
```

```python
#Mean of runs scored - rs - Chicago - since 2005
df.loc[(df.Team == 'CHC') & (df.Year > 1971)].RS.mean()
```

```python
# Exercise - What is mean of RA by Chicago until 2007?
df.loc[(df.Team == 'CHC') & (df.Year < 2007)].RA.mean()
```

##Mode

```python
#Most frequent number
#Comparing Mean Median and Mode - Wins
print(df.W.mean())
print(df.W.median())
print(df.W.mode())
```

```python
#Comparing Mean Median and Mode - OBP (On base percentage)
print(df.OBP.mean())
print(df.OBP.median())
print(df.OBP.mode())
```

```python
#Exercise: What is the OBP during the year 2010?
df.loc[df.Year == 2010].OBP.mode()
```

##Median

```python
#Central number
#Comparing Mean, Mode and Median
print(df.W.mean())
print(df.W.mode())
print(df.W.median())
```

```python
#Exercise - What is median of wins of Baltimore (BAL) until 2000?
df.loc[(df.Team == 'BAL') & (df.Year <= 2000)].W.median()
```
